```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error,r2_score

data = pd.read_csv('dataset_3.csv')


X = data.drop('MaxTemp', axis=1)  
y = data['MaxTemp']  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

knn = KNeighborsRegressor(n_neighbors=3)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", mse)
r2 = r2_score(y_test, y_pred)
print("R2 score:", r2)
```

A score of 0.99 indicates a high accuracy for your KNN classifier. This means that the classifier is correctly predicting the class labels for the majority of the samples in the test set.

```python
import matplotlib.pyplot as plt
y_pred = knn.predict(X_test)

# Plot the predicted values vs. the actual values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('KNN regression results')
plt.show()
```

The target variable wich is the Mean temperature is continous. In KNN classification, the target variable should be categorical, with a finite set of possible values. However, if the target variable is continuous, with an infinite set of possible values, then KNN classification cannot be used. Due to this reason precision, recall, and F1 score can not be calculated for this dataset.

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': range(3, 11)}

# Create a KNN regression object
knn = KNeighborsRegressor()

# Create a GridSearchCV object
grid_search = GridSearchCV(knn, param_grid, cv=5)

# Fit the GridSearchCV object to the data
grid_search.fit(X, y)

# Print the best value of K and the corresponding score
print("Best K value:", grid_search.best_params_['n_neighbors'])
print("Best score:", grid_search.best_score_)
```
